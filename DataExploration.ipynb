```python
# Import modules seciotn

import pandas as pd
import cv2
import os
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
```

```python
# Define image file and depths file dir.
current_dir = os.getcwd()
train_image_dir = current_dir + '/data/train/images/'
label_image_dir = current_dir + '/data/train/masks/'
```

```python
# Get all train image files name
train_files_list = []
label_files_list = []

# Fetch list file from train_image_dir
for root, dirs, files in os.walk(train_image_dir, topdown = False):
    train_files_list = files

# Fetch list file from test_image_dir
for root, dirs, files in os.walk(label_image_dir, topdown = False):
    label_files_list = files
    

# Sort train images file list
train_files_list.sort
label_files_list.sort

# Check number of train images. We have 4000 images for training data set.
print('Number of train images is ' + str(len(train_files_list)))
print('Number of label images is ' + str(len(label_files_list)))
```

```python
# load images with cv2
train_img_list = []
label_img_list = []

# Set progress bar
print("Download train images")
for i in tqdm(range(len(train_files_list))):
    train_img_list.append(cv2.cvtColor(cv2.imread(train_image_dir + train_files_list[i]), cv2.COLOR_BGR2RGB))
    
# Set progress bar
print("Download label images")
for i in tqdm(range(len(label_files_list))):
    label_img_list.append(cv2.cvtColor(cv2.imread(label_image_dir + label_files_list[i]), cv2.COLOR_BGR2RGB))
    
# check number of images when we load images into program
print('Number of train images loaded is ' + str(len(train_img_list)))
print('Number of label images loaded is ' + str(len(label_img_list)))
```

```python
# plot some image and check them
nrows = 2
ncols = 3
plt.figure(figsize=(24, 24), dpi=80)
for i in range(nrows * ncols):
    plt.subplot(3, 3, i + 1)
    if i < 3:
        plt.imshow(train_img_list[i])
        plt.xlabel(train_files_list[i])
    else:
        plt.imshow(label_img_list[i-ncols])
        plt.xlabel(label_files_list[i-ncols])
```

```python
# Check image_size
img_size = []
for img in train_img_list:
    height, width, channels = img.shape
    img_size_str = str(height) + 'x' + str(width) + 'x' + str(channels)
    img_size.append(img_size_str)
    
set_img_size = set(img_size)
print('All resulution of images is')
print(set_img_size)
```
